```python
#focus on first twenty-five years years
start_date = "01-01-1985"
end_date = "01-01-2010"
mask = (data.index >= start_date) & (data.index <= end_date)
data = data.loc[mask]
```

```python
data.plot()
plt.show()
```

```python
# using rolling window of twelve in order to have yearly values
yearly_data = data.rolling(window=12).mean()
data.plot()
plt.show()
```
